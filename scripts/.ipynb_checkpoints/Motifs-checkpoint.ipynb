### For consensus data


```bash
#!/bin/bash
cd ../data/motifs/
for file  in *.txt; do grep -A  30 "Full list of subgraphs size 3 ids:" $file  | tail -n +5 | cut -f 1,2,5 | sed 's/ID//g' | sed 's/ZSCORE//g'| sed '/^\s*$/d'   | awk '$3<=0.05{print $0}'> $file.Pvalue.txt; done
mv *.Pvalue.txt ../../motifs_out/
cd ../data/motifs_out/
find  -size 0 -delete
tail -n +1  *.Pvalue.txt> toTable.motifs.txt
sed 's/==>//g' toTable.motifs.txt  | sed 's/_consensus.txt.Pvalue.txt <==//g' | sed 's/_/ /g' > toTable.motifs.toexcel.txt

```

````python
consensus =pd.read_table("../data/Motifs_consensus.txt",index_col=0)
positives =pd.read_table("../data/Motifs_postitives.txt",index_col=0)
negatives =pd.read_table("../data/Motifs_negatives.txt", index_col=0)
````

## 3. Visualize dataframe 

````python
consensus
````

````python
#Normalization by motif ocurrence across levels 
consensus= consensus/consensus.sum()
consensus
````

````python
from numpy import median
ax=consensus.T.plot(kind='bar',figsize=(12,4.5), colors=colors_)
#ax=sitea.T.plot(kind='bar',figsize=(12,4))
plt.ylab=("Network motif identifier")
plt.title("Consensus Networks", weight="bold", size="large")
plt.xticks(rotation = 360)
#plt.ylim(0,5000000)
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.05 ,labelspacing=0.25)
plt.axvspan(5.5,6.5, alpha=0.1, color='salmon')
plt.tight_layout()
plt.ylabel("Motifs over-representation", size='small')
plt.savefig("../figures/Motifs_consensus.png",dpi=600,bbox_inches='tight')
````

````python
consensus= consensus/consensus.sum()
positives= positives/positives.sum()
negatives= negatives/negatives.sum()
consensus
````
