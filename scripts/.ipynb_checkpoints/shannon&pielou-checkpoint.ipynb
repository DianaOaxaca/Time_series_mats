# Read Data 

```python
shanon_genus=pd.read_table("../data/Shanon_genus.txt", index_col=0)
shanon_pfam=pd.read_table("../data/Shanon_pfam.txt",index_col=0)
pielou_genus=pd.read_table("../data/Pielou_genus.txt",index_col=0)
pielou_pfam=pd.read_table("../data/pielou.pfam.txt", index_col=0)
shanon_genus.head()
```

# Figures 

# Shanon index taxonomic level  

```python
sns.set(font_scale=1.5)
sns.set_style("darkgrid")
ax1=shanon_genus.T.plot(linewidth=1.5, linestyle=':', marker='o',label='web',figsize=(7,4),rot=40 )
plt.xticks(size='small')
plt.yticks(size='small')
plt.xlabel("Sampling time", weight='bold', size="medium" )
plt.ylabel("Shannon index",weight='bold', size="medium")
plt.ylim(4,7)
ax1.set_xlim(-.1,3.1)
ax1.xaxis.set_major_formatter(plt.NullFormatter())
plt.tight_layout()
plt.show()
#plt.savefig("../figures/Shannon.genus.time.png",dpi=400)
```

# Shanon index metabolic level  

```python
sns.set(font_scale=1.5)
sns.set_style("darkgrid")
ax1=shanon_pfam.T.plot(linewidth=1.5, linestyle=':', marker='o',label='web',figsize=(7,4),rot=40 )
plt.xticks(size='small')
plt.yticks(size='small')
plt.xlabel("Sampling time", weight='bold', size="medium" )
plt.ylabel("Shannon index",weight='bold', size="medium")
plt.ylim(7,8)
ax1.set_xlim(-.1,3.1)
#Hide the axis
ax1.xaxis.set_major_formatter(plt.NullFormatter())
plt.tight_layout
plt.show()
#plt.savefig("Shannon.pfam.time.png",dpi=500)
```

# Pielou Index Taxonomic level 


```python
ax1=pielou_genus.T.plot(linewidth=1.5, linestyle=':', marker='o',label='web',figsize=(7,3),rot=40 )
plt.xticks(size='small')
plt.yticks(size='small')
plt.xlabel("Sampling time", weight='bold', size="medium" )
plt.ylabel("Pielou index",weight='bold', size="medium")
plt.ylim(.65,.85)
plt.tight_layout()
ax1.set_xlim(-.1,3.1)
#Hide the axis
ax1.xaxis.set_major_formatter(plt.NullFormatter())
#plt.savefig("Pielou.genus.time.png",dpi=500)
```
